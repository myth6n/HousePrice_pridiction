
## **3. Feature Engineering**


**3.1 Handling Missing values**

```python
data1 = df.copy()


# /continuous features with missing values/
cont_feature = [clm for clm in missing_clmns if data1.dtypes[clm]!="O" and clm!= 'SalePrice' ] 

###categorical features with missing values ###
catg_feature = [clm for clm in missing_clmns if data1.dtypes[clm]=="O"]

```

```python
#To solve categorical features with missing values

# separating the NAN-categorical features into with and w/o meaningful values (refer dataset description)
catg_feature_with = ['FireplaceQu','GarageFinish','GarageQual','GarageCond','GarageType','BsmtExposure',
                     'BsmtCond','BsmtQual','BsmtFinType2','BsmtFinType1','MiscFeature', 'Fence', 'PoolQC','Alley']
catg_feature_w_o = ['MasVnrType','MSZoning','Functional','Utilities','Electrical','Exterior2nd','Exterior1st',
                    'KitchenQual','SaleCondition','PavedDrive','Street','LandContour', 'LotConfig','LotShape',
                    'LandContour','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType',
                    'HouseStyle','RoofStyle','RoofMatl','ExterQual','ExterCond','Foundation','Heating','CentralAir',
                    'HeatingQC','SaleType']
# 1. Impute NAn values with the constant - if it has some meaning (refer dataset description)
for feature in catg_feature_with:
    data1[feature]= data1[feature].fillna("Missing")

# 2. Impute NAn values with the Mode - if it has no meaning (refer dataset description)
for feature in catg_feature_w_o:
    data1[feature]= data1[feature].fillna(data1[feature].mode()[0])
    
```

```python
cont_w_o_NA = data1.select_dtypes(np.number)
# cont_w_o_NA = cont_w_o_NA.drop(columns=cont_feature)
# cont_w_o_NA.drop(['Id','SalePrice'],axis=1,inplace=True)
cont_w_o_NA
```

```python

#To solve continuous features with missing values-using KNN regression imputation
def knn_reg(dataset,target_na):
    cont_w_o_NA = dataset.select_dtypes(np.number)
    cont_w_o_NA = cont_w_o_NA.drop(columns=cont_feature)
    clms = cont_w_o_NA.columns
    
    cont_w_o_NA[target_na] =dataset[target_na]

    x_train = cont_w_o_NA.loc[cont_w_o_NA[target_na].isnull()==False,clms]
    
    y_train = cont_w_o_NA.loc[cont_w_o_NA[target_na].isnull()==False,target_na]
    
    x_test = cont_w_o_NA.loc[cont_w_o_NA[target_na].isnull()==True,clms]
    
    knn=KNeighborsRegressor()
    knn.fit(x_train,y_train)
    y_predict = knn.predict(x_test)
   
    dataset.loc[dataset[target_na].isnull()!=False,target_na]=y_predict
    return dataset



for feature in cont_feature:
    data2 =knn_reg(dataset=data1,target_na=feature)
```

## **4. Feature Transformation**

```python
data3= data2.copy()
```

```python
data4=data3.select_dtypes(np.number)
new_df_t = pd.DataFrame(data4.columns,columns=['feature'])
new_df_t['skewed']=scipy.stats.skew (data4)
new_df_t['abs_skewed']=abs(new_df_t['skewed'])
new_df_t['check']=new_df_t['abs_skewed'] >= 0.5
columns_t= new_df_t[new_df_t['check']==True]['feature']
```

```python
for clms in columns_t:
    data3[clms] = np.log1p(data4[clms])
```
